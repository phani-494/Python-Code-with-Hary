## match case in python

```python
# here the below program doesn't work since the python versions which have 3.10 only this match case will work 
## so now we have 3.9.4 . so the error occured.
```

```python
x = int(input("Enter the value of x: "))
# x is the variable to match
match x:
    # if x is 0
    case 0:
        print("x is zero")
    # case with if-condition
    case 4:
        print("case is 4")

    case _ if x!=90: # _ is the default one here.
        print(x, "is not 90")
    case _ if x!=80:
        print(x, "is not 80")
    case _:
        print(x)
```

```python
colors = ["Red", "Green", "Blue", "Yellow"]
for color in colors:
  print(color)
  for i in color:
    print(i)
  
for k in range(1, 12, 3):
  print(k)
```
